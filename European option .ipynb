```python
def ECC_Calculator (u, d, r, s_0, K):

    # Calculates p* and attaches the probability to each w[i]
    
    p_star = (1+r-d)/(u-d)
    probability_vector = [p_star**2, p_star*(1-p_star),p_star*(1-p_star), (1-p_star)**2]
    
    s1_u=s_0*u
    s1_d = s_0*d
    s2_uu=s_0 *u**2
    s2_ud=s_0*u*d
    s2_dd = s_0*d**2
    
    # Determines the stock tree
    
    stock_tree=[s_0, s1_u, s1_d, s2_uu, s2_ud, s2_ud, s2_dd]
    
    # Determines the option tree
    
    X= []
    option_type = input('Is the option a call or a put?')
    if option_type == 'call':
        
        for final_stock_values in stock_tree[3:7]:
            X.append(max(final_stock_values-K, 0))
    else:
        for final_stock_values in stock_tree[3:7]:
            X.append(max(K-final_stock_values, 0))
            
    # Calculates the expected value and hence the arbitrage free initial price
            
    expected_value = 0
    for i in range(len(X)):
        expected_value+=X[i]*probability_vector[i]
    expected_value = expected_value/(1+r)**2
    
# Calculates the alpha and beta 2 star values on the upper and lower branches 
    phi_2 = []
    index = 0
    for j in range(int(len(X)/2)):
        alpha_2_star = (X[index]-X[index+1])/((u-d)*stock_tree[j+1])
        phi_2.append(alpha_2_star)
        beta_2_star= (u*X[index+1]-d*X[index])/((u-d)*(1+r)**2)
        phi_2.append(beta_2_star)
        index += 2
        
        
# Calculates the V-Up and V-down values

    v_1 = [phi_2[0]*stock_tree[1] + phi_2[1]*(1+r),
     phi_2[2]*stock_tree[2] + phi_2[3]*(1+r)]

# Uses these V-Values to calculate the alpha and beta 1 star values
    
    alpha_1_star = (v_1[0]-v_1[1])/((u-d)*stock_tree[0])
    beta_1_star = (u*v_1[1]-d*v_1[0])/((u-d)*(1+r))
    phi_1 = (alpha_1_star, beta_1_star)
    
    
# Determines the trading strategy if stock is priced 1 dollar below arbitrage free initial price
    phi_1= list(phi_1)
    phi_2 = list(phi_2)
    
    psi_1_below = tuple([-phi_1[0], -phi_1[1]+1, 1])
    psi_2_below = tuple([[-phi_2[0], -phi_2[1]+1, 1], [-phi_2[2], -phi_2[3]+1, 1]])

# Determines the trading strategy if stock is priced 1 dollar above arbitrage free initial price

    psi_1_above = (phi_1[0], phi_1[1]+1, -1)
    psi_2_above = tuple([[phi_2[0], phi_2[1]+1, 1], [phi_2[2], phi_2[3]+1, 1]])
    
#Prints the data in a table 

    print( 'p* is ', p_star, '\nThe Stock Tree is ', stock_tree, '\nThe Option Tree is ', X, '\nThe Arbitrage Free Initial Price is ',expected_value,
    
    '\nThe Replicating Strategy is ', [tuple(phi_1), tuple(phi_2)],
    
     '\nIf the option is initially priced 1 dollar below the AFIP then The Trading Strategy is ', '\n',
     psi_1_below, psi_2_below,
    
    '\nOtherwise, if the option is priced 1 dollar above then The Trading Strategy is  ', psi_1_above, psi_2_above
    
    )    
    

    
    
    
    
```
